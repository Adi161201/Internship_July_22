```python
splitted_vectorized = []

for items in range(0 ,len(main_data)):
  # print(main_data['_raw'][items])
  
  try:
    temp = [main_data['_messagetimems'][items],main_data['_messagetime'][items],main_data['_size'][items],main_data['_raw'][items].split(',')]
    # _raw_splitted_vectorized.append(main_data['_raw'][items].split(''))
    splitted_vectorized.append(temp)
  except:
    continue
```

```python
main_data['_raw'][0].split(',')
```

```python
col_set = set()
for items in splitted_vectorized:
  # print(items)
  try:
    for label in items[3]:
      something = label.split('=')
      if len(something) >= 2:
        if ':' not in something[0] and '.' not in something[0] and '}' not in something[0] and '{' not in something[0]:
          col_set.add(something[0].strip())
  except:
    continue
```
